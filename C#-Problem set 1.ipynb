```
using System;

namespace ConsoleApp3
{
    public class Program
    { 
    static bool Inside(double x1, double y1, double x2,
                         double y2, double x, double y)
    {
        if (x > x1 && x < x2 &&
            y > y1 && y < y2)
            return true;

        return false;
    }

    
    public static void Main()
    {


        //first rectangle
        double x1 = 0.3, y1 = 0.5, x2 = 1.1, y2 = 0.7;

        //second rectangle
        double x11 = 0.5, y11 = 0.2, x22 = 1.1, y22 = 2.0;

        
        Console.WriteLine("Enter value for x:");
        int x = Convert.ToInt32(Console.ReadLine());

        Console.WriteLine("Enter value for y:");
        int y = Convert.ToInt32(Console.ReadLine());


        
        if (Inside(x1, y1, x2, y2, x, y))
            Console.Write("B.1 - True\n");
        else
            Console.Write("B.1 - False\n");

        if (Inside(x11, y11, x22, y22, x, y))
            Console.Write("B.2 - True\n");
        else
            Console.Write("B.2 - False\n");
    }

}

}
```
